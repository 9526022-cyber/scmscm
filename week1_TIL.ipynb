## 전화 지역번호와 지역에 대한 딕셔너리 만들기

```python
phone = ">경기 031 >강원 033 >충남 041 >충북 043 >경북 054 >경남 055 >전남 061 >전북 063"
phone
```

```python
phone_list = phone.split(">")
phone_list
```

```python
phone_list[1].split()
```

```python
phone_dict = {}
for p in phone_list:
    p_split = p.split()
    if len(p_split) > 0:
        key = p_split[0]
        val = p_split[1]
        phone_dict[key] = val
        print(key,val)
```

```python
phone_dict
```
